```
# Separating input and output parameters
x = data.drop("income", axis=1)
y = data["income"]
```

```
x = pd.get_dummies(x, drop_first=True)
y = y.apply(lambda val: 1 if val.strip() == ">50K" else 0)
```

```
# Splitting dataset for Training and Testing 80/20
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=59, stratify=y)
```

```
# Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```
# Logistic Regression
from sklearn.linear_model import LogisticRegression

logReg = LogisticRegression()
logReg.fit(x_train, y_train)
```

```
y_predLR = logReg.predict(x_test)
```

```
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predLR, target_names=["<=50k", ">50k"]))

cmLR = confusion_matrix(y_test, y_predLR)
print(f"Confusion Matrix:\n {cmLR}\n")

sea.heatmap(cmLR, annot=True, fmt="d", cmap="RdYlGn", cbar=False, xticklabels=["<=50K", ">50K"], yticklabels=["<=50K", ">50K"])
plot.xlabel("Predicted")
plot.ylabel("Actual")
plot.title("Confusion Matrix")
plot.show()
```

```
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

decTre = DecisionTreeClassifier(max_depth=10, min_samples_split=5, criterion="entropy", random_state=59)
decTre.fit(x_train, y_train)
```

```
y_predDT = decTre.predict(x_test)
```

```
cmDT = confusion_matrix(y_test, y_predDT)

sea.heatmap(cmDT, annot=True, fmt="d", cmap="RdYlGn", cbar=False, xticklabels=["<=50K", ">50K"], yticklabels=["<=50K", ">50K"])
plot.xlabel("Predicted")
plot.ylabel("Actual")
plot.title("Confusion Matrix")
plot.show()
```

```
from sklearn.tree import plot_tree

plot.figure(figsize=(20,10))
plot_tree(decTre, filled=True, feature_names=x.columns, class_names=["<=50K", ">50K"], max_depth=2)
plot.show()
```

```
print(classification_report(y_test, y_predDT))
```

```
# Random Forest
from sklearn.ensemble import RandomForestClassifier

ranFor = RandomForestClassifier(n_estimators=200, max_depth=10, min_samples_split=5, random_state=59)
ranFor.fit(x_train, y_train)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Input
from scikeras.wrappers import KerasClassifier

def build_keras_model():
    neuNet = Sequential([
        Input(shape=(x_train.shape[1],)),
        Dense(128, activation="relu"),
        Dropout(0.4),
        Dense(64, activation="relu"),
        Dropout(0.3),
        Dense(32, activation="relu"),
        Dense(1, activation="sigmoid")    ])

    neuNet.compile(optimizer=Adam(learning_rate=0.0005), loss="binary_crossentropy", metrics=["accuracy"])
    return neuNet


keras_model = KerasClassifier(model=build_keras_model, epochs=50, batch_size=32, verbose=0)
keras_model.fit(x_train, y_train)
```

```
y_predNN = keras_model.predict(x_test)
```

```
cmNN = confusion_matrix(y_test, y_predNN)

sea.heatmap(cmNN, annot=True, fmt="d", cmap="RdYlGn", cbar=False, xticklabels=["<=50K", ">50K"], yticklabels=["<=50K", ">50K"])
plot.xlabel("Predicted")
plot.ylabel("Actual")
plot.title("Confusion Matrix")
plot.show()
```

```
print(classification_report(y_test, y_predNN))
```

```
from sklearn.ensemble import StackingClassifier

estimators = [
    ("lr", logReg),
    ("rf", ranFor),
    ("gbm", gbm),
    ("svm", svm),
    ("nn", keras_model),
]
```

```
from sklearn.metrics import accuracy_score

stacked_model = StackingClassifier(
    estimators=estimators,
    final_estimator=GradientBoostingClassifier(),
    passthrough=False,
    cv=5
)

stacked_model.fit(x_train, y_train)

y_predST = stacked_model.predict(x_test)

print(f"Stacked model Accuracy: {accuracy_score(y_test, y_predST):.4f}")
print(classification_report(y_test, y_predST))
```

```
cmST = confusion_matrix(y_test, y_predST)

sea.heatmap(cmST, annot=True, fmt="d", cmap="RdYlGn", cbar=False, xticklabels=["<=50K", ">50K"], yticklabels=["<=50K", ">50K"])
plot.xlabel("Predicted")
plot.ylabel("Actual")
plot.title("Confusion Matrix")
plot.show()
```

```
%%writefile app.py

import streamlit as st
import pandas as pd
import numpy as np
import joblib
import pickle
from tensorflow.keras.models import load_model

model = load_model("keras_income_model.h5")
scaler = joblib.load("scaler.pkl")

with open("columns.pkl", "rb") as f:
    model_columns = pickle.load(f)

education_map = {'Preschool': 1, '1st-4th': 2, '5th-6th': 3, '7th-8th': 4, '9th': 5, '10th': 6,
                 '11th': 7, '12th': 8, 'HS-grad': 9, 'Some-college': 10, 'Assoc-voc': 11,
                 'Assoc-acdm': 12, 'Bachelors': 13, 'Masters': 14, 'Prof-school': 15, 'Doctorate': 16 }

st.set_page_config(page_title="Income Predictor", layout="centered")
st.markdown("<h1 style='text-align: center;'>💼 Salary Prediction Using Machine Learning</h1>", unsafe_allow_html=True)
st.markdown("<p style='text-align: center; font-size:17px;'>Classifying salaries as more or less than $50K/year</p>", unsafe_allow_html=True)
st.markdown("---")
st.markdown("<br>", unsafe_allow_html=True)

with st.form("income_form"):
    st.markdown("### 👤 Personal Details")

    col1, spacer, col2 = st.columns([1, 0.3, 1])

    with col1:
        age = st.slider("🎂 Age", 18, 75, 30)
        education = st.selectbox("🎓 Education", list(education_map.keys()))
        marital_status = st.selectbox("💍 Marital Status", [
            'Never-married', 'Married-civ-spouse', 'Divorced',
            'Separated', 'Widowed', 'Married-spouse-absent'
        ])
        relationship = st.selectbox("👨‍👩‍👧 Relationship", [
            'Wife', 'Own-child', 'Husband', 'Not-in-family',
            'Other-relative', 'Unmarried'
        ])
        hours_per_week = st.slider("⏱️ Hours per Week", 1, 100, 40)

    with col2:
        workclass = st.selectbox("🏢 Workclass", [ "Private", "Self-emp-not-inc", "Self-emp-inc", "Federal-gov",
                                                  "Local-gov", "State-gov", "Without-pay", "Never-worked", 'Others'])

        occupation = st.selectbox("💼 Occupation",
                                  ["Tech-support", "Craft-repair", "Other-service", "Sales", "Exec-managerial",
                                  "Prof-specialty", "Handlers-cleaners", "Machine-op-inspct", "Adm-clerical",
                                  "Farming-fishing", "Transport-moving", "Priv-house-serv", "Protective-serv", "Armed-Forces", "Others"])

        race = st.selectbox("🌎 Race", [
            "White", "Black", "Asian-Pac-Islander", "Amer-Indian-Eskimo", "Other" ])

        gender = st.radio("⚧️ Gender", ["Male", "Female"], horizontal=True)

        native_country = st.selectbox("🌐 Region", [
            "North America", "Central America", "South America",
            "Europe", "Asia", "Other" ])

    st.markdown("<br>", unsafe_allow_html=True)
    submitted = st.form_submit_button("🔍 Predict")


if submitted:
    st.markdown("<br><hr><br>", unsafe_allow_html=True)
    st.markdown("### 📋 Input Summary")

    education_num = education_map[education]

    input_data = {
        "age": age,
        "workclass": workclass,
        "educational-num": education_num,
        "marital-status": marital_status,
        "occupation": occupation,
        "relationship": relationship,
        "race": race,
        "gender": gender,
        "hours-per-week": hours_per_week,
        "native-country": native_country
    }

    input_df = pd.DataFrame([input_data])
    st.dataframe(input_df, use_container_width=True)

    st.markdown("<br>", unsafe_allow_html=True)

    df_encoded = pd.get_dummies(input_df)
    df_encoded = df_encoded.reindex(columns=model_columns, fill_value=0)
    df_scaled = scaler.transform(df_encoded)

    prediction = model.predict(df_scaled)
    prob = float(prediction[0][0])
    label = ">50K" if prob > 0.5 else "<=50K"

    st.markdown("### 🎯 Prediction Result")
    st.success(f"📊 Predicted Income: **{label}**")
    st.markdown(f"✅ Confidence: `{prob:.2%}`")
    st.progress(prob)

```
